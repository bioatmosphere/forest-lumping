```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

## Fig 1a

```python
biomass_by_species = pd.read_csv('fig_1a.csv',index_col=0)
biomass_by_species.T.head()
```

```python
biomass_by_species.index
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,6))

new_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728',
              '#9467bd', '#8c564b', '#e377c2', '#7f7f7f',
              '#bcbd22', '#17becf', 'b',       'grey',
              'royalblue','tomato']

ax.stackplot(biomass_by_species.index[:],biomass_by_species.T.iloc[:,:],
             labels=biomass_by_species.columns,
             colors=new_colors)

ax.tick_params(bottom=True, labelsize=14)
ax.set_xlabel('Year of Simulation',size=14)
ax.set_ylabel('Total biomass(Mg C ha$^-$$^1$)',size=14)
ax.set_xlim(-10,510)
ax.set_ylim(-5, 185)

ax.legend(frameon=False,ncol=4,fontsize=10)

plt.savefig('fig_1a.pdf',bbox_inches='tight')

plt.show()
```

## Fig 1b

```python
mass_lump_method = pd.read_csv('fig_1b.csv',index_col=0)
mass_lump_method.head()
```

```python
mass_lump_method.columns[0]
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,4))

#ax.plot(mass_lump_method.loc[:,'year'],mass_lump_method.iloc[:,1:])
ax.plot(mass_lump_method.iloc[:,0],color='grey',label='Control',linewidth=2)
ax.plot(mass_lump_method.iloc[:,1],label='One PFT',linewidth=2)
ax.plot(mass_lump_method.iloc[:,2],label='Evergreenness',linewidth=2)
ax.plot(mass_lump_method.iloc[:,3],label='Tolerance',linewidth=2)
ax.plot(mass_lump_method.iloc[:,4],label='Genus',linewidth=2)


ax.set_xlabel('Year of Simulation',size=14)
ax.set_ylabel('Total biomass(Mg C ha$^-$$^1$)',size=14)
ax.tick_params(bottom=True, labelsize=14)
ax.set_xlim(-10,510)
ax.set_ylim(-5, 205)

ax.legend(frameon=False,fontsize=14)

plt.savefig('fig_1b.pdf',bbox_inches='tight')
plt.show()
```

```python
tol_by_mass = pd.read_csv('fig_2.csv',index_col=0)
tol_by_mass.head()
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,7))

labels = tol_by_mass.columns

colors=['purple','orange','green','black','red','blue']


for i in range(6):
    ax.plot(tol_by_mass.iloc[:,i],label=labels[i],color=colors[i],linewidth=2)
    
for i in range(6,11):
    ax.plot(tol_by_mass.iloc[:,i],label=labels[i],linestyle='--')

ax.plot(mass_lump_method['Control'],color='grey',label='Control',linewidth=2)

ax.set_xlabel('Year of Simulation',size=14)
ax.set_ylabel('Total biomass(Mg C ha$^-$$^1$)',size=14)
ax.tick_params(bottom=True, labelsize=14)
ax.set_xlim(-10,510)
ax.set_ylim(-5, 320)

ax.legend(ncol=3,frameon=False,fontsize=14)

plt.savefig('fig_2.pdf',bbox_inches='tight')

plt.show()
```
